```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("D:\Pranav\CL_4\BI\Datasets\Global_Superstore2.csv", encoding='latin1')
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df.drop_duplicates()
```

```python
# date-time conversion
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df['Month'] = df['Order Date'].dt.to_period('M')
df['Shipping Delay (Days)'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
df.to_csv('cleaned_superstore.csv')
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(12, 6))    # width and height of the plot in inches
monthly_sales = df.groupby('Month')['Sales'].sum()  # grops data by month col aggregates or sums the sales values for each month giving total monthly sales
monthly_sales.plot(kind='line', marker='o')   # plots in line chart, marker='o' puts a dot on each data point, making the trend easier to observe
plt.title('Monthly Sales Trend')
plt.ylabel('Sales ($)')
plt.xticks(rotation=45)  # Rotates the X-axis (Month) labels by 45 degrees to avoid overlap and make it easier to read.
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))
region_profit = df.groupby('Region')['Profit'].sum().sort_values()   # groups dataset by rehion col, agggregates the profit col by sum and sorts ele in asc order
sns.barplot(x=region_profit.index, y=region_profit.values, palette='coolwarm')
plt.title('Total Profit by Region')
plt.ylabel('Profit ($)')
plt.xlabel('Region')
plt.tight_layout()   #ensures the plot elements are not cut off
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Category')  # hue-Uses different colors for different product categories
plt.title('Sales vs. Profit by Category')
plt.tight_layout()
plt.show()
```
